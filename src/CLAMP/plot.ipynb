```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

BERT_CLAMP_40

```python
df_train = pd.read_csv("/home/hieunghiem/CHSI/MB_BERT_new/CLAMP/Train_Eval_loss/training_loss_bert.csv", sep=",", header=None)
df_val = pd.read_csv("/home/hieunghiem/CHSI/MB_BERT_new/CLAMP/Train_Eval_loss/validation_loss_bert.csv", sep=",", header=None)

fig, axs = plt.subplots(1, 5, figsize=(20, 4))
fig.suptitle('BD-BERT with CLAMP - 40 epochs', fontsize=17, y=1)

for j in range(5):

    epochs = np.arange(1, 41)

        # Plotting training and validation losses
    axs[j].plot(epochs, df_train.iloc[j].to_list(), label='Training Loss')
    axs[j].plot(epochs, df_val.iloc[j].to_list(), label='Validation Loss')
    # Setting x-axis and y-axis ticks
    axs[j].set_xticks(np.arange(0, 41, 5))  # X-axis ticks from 0 to 20 with step 5
    axs[j].set_yticks(np.arange(0, 1.3, 0.2))  # Y-axis ticks from 0 to 1 with step 0.1


    axs[j].legend()
    axs[j].set_title(f"Fold {j}")

        # Display the plot


plt.tight_layout()
plt.subplots_adjust(wspace=0.3)

plt.show()
```
